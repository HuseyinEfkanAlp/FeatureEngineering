# Missing values (Eksik Degerler)

* Gozlemler de eksiklik olmasi durumunu ifade eder

## Eksik değer problemi nasıl çözülür?

* Silme
* Değer Atama (Mod, Median)
* Tahmine Dayali Yontemler (Makine ogrenmesi, Istatistiksel yontemler)

######### Eksik veri ile calisirken goz onunde bulundurulmasi gereken onemli konulardan birisi: *Eksik verinin rassalligi*
Eksik verilerin rastgele ortaya cikip cikmadigi durumu

Eksik degere sahip gozlemlerin veri setinden dirket cikarilmasi ve rassalligin incelenmemesi, yapilacak istatistiksel cikarimlarin ve modelleme calismalarinin guvenilirligini dusurecektir.(Reha Alpar, 2011) 

Eger eksiklikler degiskenler ile iliskili olarak ortaya cikan yapisal problemler ile meydana gelmis ise bu durumda yapilacak silme islemleri ciddi yanliliklara sebep olabilecektir. (Tabachnick ve Fidell, 1996)

```python
### Eksik gozlem var mi yok mu?
# isnull = butun df i gez eksiklik varmi bak True/False
# values = bu true false lari tut
# any = herhangi bir true varsa true yaz
df.isnull().values.any()
```

```python
# Veri setindeki toplam eksik deger sayisi
df.isnull().sum().sum()
```

```python
# En az bir tane eksik degere sahip olan gozlem birimleri
df[df.isnull().any(axis=1)]
```

```python
# Tam olan eksik olmayan gozlem birimlerini getir
df[df.notnull().any(axis=1)]
```

```python
# Eksikligin butun veri setindeki orani
(df.isnull().sum()/ df.shape[0]*100).sort_values(ascending= False)
```

```python
def missing_values_table(df, na_name=False):
    na_columns = [col for col in df.columns if df[col].isnull().sum()>0]
    n_miss = df[na_columns].isnull().sum().sort_values(ascending=False)
    ratio = (df[na_columns].isnull().sum()/df.shape[0]*100).sort_values(ascending=False)
    missing_df = pd.concat([n_miss, np.round(ratio, 2)], axis=1, keys=['n_miss', 'ratio'])
    print(missing_df, end='\n')
    if na_name:
        return na_columns
```

```python
missing_values_table(df)
```
